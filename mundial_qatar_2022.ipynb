### Segunda Parte

```python
equipo : str = "ARGENTINA"
```

```python
# Seleccionar todos los partidos que jugo el equipo de "local"
# Completar
df : pd.DataFrame = pd.read_csv(archivo)
ARG_local : pd.Series = df["1"] == equipo
df[ARG_local]

```

```python
# Seleccionar todos los partidos que jugo el equipo de "visitante"
# Comp
ARG_visita : pd.Series = df["2"] == equipo
df[ARG_visita]
```

```python
# Seleccionar todos los partidos que jugo el equipo (de "local" o "visitante")
# Completar

ARG_partidos : pd.DataFrame = (df["1"] == equipo) | (df["2"] == equipo)
df[ARG_partidos]
```

```python
# Cuantos goles hizo el equipo en total (obtener los goles metidos de local y sumarlos con los goles metidos de visitante)
# Completar

golesLocal : int = df[ARG_local]["1_goals"].sum()
golesVisita : int = df[ARG_visita]["2_goals"].sum()
goles : int = golesLocal + golesVisita
goles
```

```python
cant_partidos_arg = len(df[ARG_local | ARG_visita])
```

```python
# Cuantos goles hizo el equipo en promedio
# Completar
goles_promedio : float= goles / cant_partidos_arg
goles_promedio
```

```python
# Cuantos partidos gano el equipo
# Completar

localGanados = (df[ARG_local]["1_goals"]) > (df[ARG_local]["2_goals"])
visitaGanados = (df[ARG_visita]["1_goals"]) < (df[ARG_visita]["2_goals"])
ganados = len(localGanados) + len(visitaGanados)
ganados





```
